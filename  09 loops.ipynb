```python
l= ["Apples", "Bananas", "Tomatoes", "Pineapples", "Guava"]
for items in l:
    if (items == "Pineapples"):
        break            # The loop breaks on the given itemset
    print(items)
```

```python
l= ["Apples", "Bananas", "Tomatoes", "Pineapples", "Guava"]
for items in l:
    if (items == "Pineapples"):
        continue         # The loop skips the particular items and print the others 
    print(items)
```
